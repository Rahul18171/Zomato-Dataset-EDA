```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country
```

```python
# merge two dataframe based on country code file
Final_df=pd.merge(df,df_country, on='Country Code',how='inner')
```

```python
Final_df.head()
```

```python
# check datatypes
Final_df.dtypes
```

```python
Final_df.columns
```

```python
# lets check country name and their counts
Final_df.Country.value_counts()
```

```python
Country_names=Final_df.Country.value_counts().index
```

```python
Country_val=Final_df.Country.value_counts().values
```

```python
#Pie Chart- Top 3 countries use zomato
plt.pie(Country_val[:3],labels=Country_names[:3],autopct='%1.2f%%')
plt.show()
```

Analysis= Zomato maximum records or transactions are from India, after that USA and United Kingdom.
 

```python
Final_df.columns
```

```python
ratings=Final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

   Analysis:
   
1.When rating is between 4.5 to 4.9 ---> Excellent

2.When rating is between 4.0 to 4.4 ---> Very Good

3.When rating is between 3.5 to 3.9 ---> Good

4.When rating is between 2.5 to 3.4 ---> Average

5.When rating is between 1.8 to 2.4 ---> Poor

```python
import matplotlib
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating Count', data=ratings)
plt.show()
```

```python
# if we want rating like rating colour  # we change here white colour its get displayed and last one also 
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color', data=ratings,palette=['Blue','Red','Orange','Yellow','Green','Purple'])
```

Analysis:
1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4 

```python
# Value Counts
plt.figure(figsize=(10,6))
sns.countplot(x='Rating color', data=ratings,palette=['Blue','Red','Orange','Yellow','Green','Purple'])
```
